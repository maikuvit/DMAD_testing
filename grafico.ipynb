```python
from matplotlib.pyplot import plot 
```

```python
line = "Finished epoch:1 | Loss : 1.99951 | Learning Rate: 0.5"
line.split('|')[1]
```

```python
# read the file diffusion_losses.txt formatted as
#  Finished epoch:1 | Loss : 1.99951 | Learning Rate: 0.5 
# and plot the loss as a function of the epoch

with open('diffusion_losses.txt') as f:
    lines = f.readlines()
    epochs = []
    losses = []
    for line in lines:
        if 'Finished epoch:' in line:
            epoch = int(line.split(' ')[1].split(':')[1])
            loss = float(line.split('|')[1].split(':')[1])
            epochs.append(epoch)
            losses.append(loss)
    plot(epochs, losses)
```
